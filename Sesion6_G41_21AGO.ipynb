```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nacimientos=pd.DataFrame({'Municipio':['Monterrey','Apodaca','Escobedo','Santiago'],'Bebes_nacidos':[233,569,780,123],'Muertos':[123,80,23,12]})
```

```python
nacimientos
```

```python
nacimientos=nacimientos.sort_values('Bebes_nacidos',ascending=False)
```

```python
plt.bar(nacimientos['Municipio'],nacimientos['Bebes_nacidos'],color=['black'])
plt.title('Nacimientos en NL en 2020')
plt.xlabel('Municipios')
plt.ylabel('Cantidad de bebes nacidos')
plt.show()
```

```python
personas=['Juan','Pedro','Ana','Ale']
edades=[23,18,20,31]
```

```python
plt.bar(personas,edades)
plt.show()
```

```python
#Lienzo
fig,(panel1,panel2)=plt.subplots(2,1)
fig.tight_layout()

#grafica1 nacimientos
x1=nacimientos['Municipio']
y1=nacimientos['Bebes_nacidos']
panel1.bar(x1,y1)

x2=personas
y2=edades
panel2.bar(x2,y2)
plt.show()
```
