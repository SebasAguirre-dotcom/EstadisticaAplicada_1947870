```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_predict
## estimar modelo ARIMA
from statsmodels.tsa.arima.model import ARIMA
import pmdarima as pm
```

```python
desempleo_mex=pd.read_excel('desempleo_mex.xlsx')
```

```python
desempleo_mex.head()
```

```python
desempleo_mex.dtypes
```

```python
desempleo_mex.isnull().sum()
```

```python
## convertir en tiempo
desempleo_mex['meses']=pd.to_datetime(desempleo_mex['meses'],format='%Y/%m')
```

```python
desempleo_mex
```

```python
desempleo_mex=desempleo_mex.set_index('meses')
desempleo_mex=desempleo_mex.asfreq('MS')
```

```python
desempleo_mex.index
```

```python
desempleo_mex['tasa de desempleo'].plot(figsize=(10,6))
plt.xlabel('mes')
plt.ylabel('%')
plt.title('Tasa de desempleo en Mexico 2005-2023')
```

```python
## Prueba dickey-fuller
prueba_adf=adfuller(desempleo_mex['tasa de desempleo'])
pvalue=prueba_adf[1]
pvalue.round(2)
```

No rechazamos H0, significa que la tasa de desempleo es un proceso no estacionario

```python
## 1ras diferencias
desempleo_mex['desempleo_diff']=desempleo_mex['tasa de desempleo'].diff()
```

```python
desempleo_mex['desempleo_diff'].plot()
```

```python
prueba_adf1=adfuller(desempleo_mex['desempleo_diff'].dropna())
pvalue1=prueba_adf1[1]
pvalue1.round(2)
```

Rechazamos H0, las primeras diferencias son un proceso estacionario

```python
ig, ax=plt.subplots(1,2,figsize=(10,4))
plot_acf(desempleo_mex['desempleo_diff'].dropna(),zero=False, lags=30, ax=ax[0])
plot_pacf(desempleo_mex['desempleo_diff'].dropna(),zero=False, lags=40, ax=ax[1])
plt.show()
```

El desempleo es una serie con componente estacional

```python
## determinar el orden del SARIMA
pm.auto_arima(desempleo_mex['tasa de desempleo'], start_p=1, max_p=7, start_q=1,max_q=7, ic='aic', m=12,seasonal=True, start_P=1, start_Q=1, max_P=7, max_Q=7)
```

El desempleo es una SARIMA (1,1,0)(1,0,1,12)

```python
#plt.rcParams['figure.figsize']=(14,4)
#plot_acf(desempleo_mex['desempleo_diff'].dropna(),zero=False, lags=30)
```

```python
model=ARIMA(desempleo_mex['tasa de desempleo'], order=(1,1,0), seasonal_order=(1,0,1,12))
results=model.fit()
print(results.summary())
```

```python
desempleo_mex.tail()
```

```python
# Pronostico
plot_predict(results, '2023-09-01','2024-02-01')
desempleo_mex['tasa de desempleo'].plot()
plt.title('Desempleo en Mexico')
```

```python
results.predict('2023-09-01','2024-02-01')
```

```python
forecast=results.get_forecast(steps=6)
forecast_mean=forecast.predicted_mean
forecast_ci=forecast.conf_int()
```

```python
inicio='2023-09-01'
fin='2024-02-01'
```

```python
pronostico_valor=forecast_mean[inicio:fin]
pronostico_intervalos=forecast_ci.loc[inicio:fin]
```

```python
df_pronostico=pd.DataFrame({'pronostico': pronostico_valor, 'limite inferior':pronostico_intervalos.iloc[:,0],'limite superior':pronostico_intervalos.iloc[:,1]})
df_pronostico
```

```python
print(f'La tasa de desempleo pronosticada para septiembre 2023 es {df_pronostico.iloc[0,0].round(2)}% pero puede estar entre {df_pronostico.iloc[0,1].round(2)}% y {df_pronostico.iloc[0,2].round(2)}%')
```
